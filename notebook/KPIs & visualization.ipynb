## Visualization Section

### 1)Top 10 most popular tracks
### 2)Top 10 artists with most popular songs
### 3)Top 7 most loud songs
### 4)Top 7 artists with danceable overall songs
### 5)Overall relationship between columns :
        ex: danceability & loudness 
        ex: valence & mode

```python
df['popularity'].max()
```

```python
df_sorted_by_popular_songs = df.drop_duplicates(subset=['track_name'], keep='first')
df_sorted_by_popular_songs = df_sorted_by_popular_songs.sort_values(by='popularity',axis=0,ascending=False).head(10).reset_index().drop(columns= 'index')
df_sorted_by_popular_songs
```

```python
colors = ['#085D70','#1A6A7F','#2C768D','#3E839C','#508FAB','#619CB9','#73A8C8','#85B5D7','#97C1E5','#A9CEF4']
Top_10_most_popular_tracks_fig = px.histogram(df_sorted_by_popular_songs,x='track_name',y='popularity',histfunc='max',color='popularity',color_discrete_sequence= colors)
Top_10_most_popular_tracks_fig
```

```python
df_sorted_by_popular_artists = df.drop_duplicates(subset=['artists'], keep='first')
df_sorted_by_popular_artists = df_sorted_by_popular_artists.sort_values(by='popularity',axis=0,ascending=False).head(10).reset_index().drop(columns= 'index')
df_sorted_by_popular_artists
```

```python
Top_10_most_popular_songs_artists_fig = px.histogram(df_sorted_by_popular_artists,x='artists',y='popularity',histfunc='max',color='popularity',color_discrete_sequence= colors)
Top_10_most_popular_songs_artists_fig
```

```python
df_sorted_by_most_loud_songs= df.sort_values(by='loudness',ascending=False).head(7).reset_index().drop(columns= 'index')
df_sorted_by_most_loud_songs
```

```python
Top_7_most_loud_songs_fig = px.histogram(df_sorted_by_most_loud_songs,x='track_name',y='loudness',histfunc='max',color='track_name',color_discrete_sequence= colors)
# Top_7_most_loud_songs_fig = px.pie(df_sorted_by_most_loud_songs,names='track_name',values='loudness',color='loudness',color_discrete_sequence= colors)
Top_7_most_loud_songs_fig
```

```python
df_sorted_by_most_dance_artists= df.sort_values(by='danceability',ascending=False).head(7).reset_index().drop(columns= 'index')
df_sorted_by_most_dance_artists
```

```python
Top_7_most_dance_artists_fig = px.pie(df_sorted_by_most_dance_artists,names='artists',values='danceability',color='danceability',color_discrete_sequence= colors)
Top_7_most_dance_artists_fig
```
